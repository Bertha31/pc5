```python
import pandas as pd

# Cargar los datos en un DataFrame
df = pd.read_excel('D:\\PYTHON\\Python-PC5-main\\src\\cripto_currency.xlsx')



```

```python
df['Tipo de Moneda'] = 'Cripto'


```

```python
resumen = df.groupby('Tipo de Moneda')['Close'].agg(['mean', 'min', 'max'])


```

```python
resumen.to_excel('reporte_resumen.xlsx')
```

```python
import matplotlib.pyplot as plt

# Genera una gráfica
plt.plot(df['Date'], df['Close'])
plt.xlabel('Fecha')
plt.ylabel('Cierre')
plt.title('Gráfico de Cierre de Criptomonedas')

# Guarda la gráfica en un archivo
plt.savefig('grafico_cierre.png')

# Agrega la imagen al archivo Excel
from openpyxl import Workbook
from openpyxl.drawing.image import Image

wb = Workbook()
ws = wb.active
img = Image('grafico_cierre.png')
ws.add_image(img, 'A10')
wb.save('reporte_resumen.xlsx')
```
